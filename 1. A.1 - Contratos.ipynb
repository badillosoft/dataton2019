```python
from pymongo import MongoClient

mongo_client = MongoClient('localhost', 27017)

db = mongo_client["dataton2019"]

coll = db["contratos{}".format(year)]

coll
```

```python
import re

def keymap(key):
    return re.sub(r"[^a-zA-Z0-9-]", "x", re.sub(r"\s+", "-", key)).lower()
```

```python
coll.drop()

for rowIndex, row in data.iterrows():
#     if rowIndex > 3:
#         break
    frame = {
        "columns": []
    }
    for colIndex, col in enumerate(row):
        key = keymap(data.columns[colIndex])
        frame["columns"].append((key, data.columns[colIndex]))
        frame[key] = str(col).strip()
        # print(rowIndex, colIndex, data.columns[colIndex], col)
    frame["_id"] = "{}-{}-{}-{}-{}".format(
        frame[keymap("Fecha de publicación")][:10], 
        frame[keymap("Orden de gobierno")],
        frame[keymap("Clave de la UC")],
        frame[keymap("Código del expediente")],
        rowIndex
    )
    coll.insert_one(frame)
    if rowIndex % 10000 == 0:
        print("...({})".format(rowIndex))
#     print(frame)
```
